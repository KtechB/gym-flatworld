```python
def get_obs_from_pkl(path):
    with open(path, "rb") as f:
        epis = pickle.load(f)
    obs = [0] * len(epis)
    for i in range(len(epis)):
        obs[i] = epis[i]["obs"]
    obs = np.reshape(np.array(obs), (-1, epis[i]["obs"][0].shape[0]))
    return obs
```

```python
obs = get_obs_from_pkl(EPIS_PATH)
```
